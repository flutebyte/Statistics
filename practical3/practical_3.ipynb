Q1) Identify all duplicate cases using prof. Using all observations, find the average and standard deviation for age. Repeat the analysis by first filtering the data set to include one observation for each instructor with a total number of observations restricted to 94.

```python
import pandas as pd
data = {
    'prof': ['Smith', 'Jones', 'Smith', 'Taylor', 'Brown', 'Taylor', 'Jones', 'Evans'],
    'age': [40, 52, 55, 88, 39, 38, 59, 69]
}
df = pd.DataFrame(data)

duplicates = df[df.duplicated('prof', keep=False)]
print("Duplicate cases based on 'prof':")
print(duplicates)

mean_all = df['age'].mean()
std_all = df['age'].std()

print("\nUsing all observations:")
print(f"Mean age = {mean_all:.2f}")
print(f"Standard deviation = {std_all:.2f}")

df_unique = df.drop_duplicates(subset='prof', keep='first')

print(f"\nNumber of unique professors: {df_unique['prof'].nunique()}")

mean_unique = df_unique['age'].mean()
std_unique = df_unique['age'].std()

print("\nUsing one observation per professor (n = 94 or less):")
print(f"Mean age = {mean_unique:.2f}")
print(f"Standard deviation = {std_unique:.2f}")
```

Q2) Using a bar chart, demonstrate if instructors teaching lower-division courses receive higher average teaching evaluations.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'course_division': ['lower', 'upper', 'lower', 'upper', 'lower', 'upper', 'lower', 'upper'],
    'evaluation': [3.8, 4.5, 4.0, 4.7, 3.9, 4.6, 3.7, 4.8]
}

df = pd.DataFrame(data)
avg_eval = df.groupby('course_division')['evaluation'].mean().reset_index()

plt.bar(avg_eval['course_division'], avg_eval['evaluation'],
        color=['skyblue', 'lightgreen'], edgecolor='black')
plt.title('Average Teaching Evaluation by Course Division')
plt.xlabel('Course Division')
plt.ylabel('Average Evaluation Score')
plt.ylim(0, 5)
plt.grid(axis='y', alpha=0.5)
plt.show()
```

Q3) Plot the relationship between age and teaching evaluation scores

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60],
    'Teaching_Evaluation': [4.6, 4.5, 4.3, 4.1, 3.9, 3.8, 3.7, 3.6]
}
df = pd.DataFrame(data)

plt.figure(figsize=(8,6))
sns.regplot(x='Age', y='Teaching_Evaluation', data=df, marker='o', color='blue')
plt.title('Relationship between Age and Teaching Evaluation Scores')
plt.xlabel('Age')
plt.ylabel('Teaching Evaluation Score')
plt.grid(True)
plt.show()
```

Q4) Using gender-differentiated scatter plots, plot the relationship between age and teaching evaluation scores.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 28, 33, 38, 42],
    'Teaching_Evaluation': [3.7, 3.8, 3.9, 4.0, 4.1, 4.0, 4.2, 4.1, 4.3, 4.4, 4.5, 4.6],
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male',
               'Female', 'Female', 'Female', 'Female']
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,6))
sns.scatterplot(
    x='Age',
    y='Teaching_Evaluation',
    hue='Gender',
    style='Gender',
    s=100,
    data=df
)
plt.title('Relationship between Age and Teaching Evaluation Scores by Gender')
plt.xlabel('Age')
plt.ylabel('Teaching Evaluation Score')
plt.grid(True)
plt.legend(title='Gender')
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 28, 33, 38, 42],
    'Teaching_Evaluation': [4.4, 4.3, 4.1, 4.0, 3.9, 3.8, 3.7, 3.6, 3.9, 4.1, 4.3, 4.5],
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male',
               'Female', 'Female', 'Female', 'Female']
}

df = pd.DataFrame(data)

sns.lmplot(
    x='Age',
    y='Teaching_Evaluation',
    hue='Gender',
    data=df,
    markers=['o', 's'],
    height=6,
    aspect=1.2
)
plt.title('Age vs Teaching Evaluation Scores by Gender')
plt.xlabel('Age')
plt.ylabel('Teaching Evaluation Score')
plt.grid(True)
plt.show()
```
